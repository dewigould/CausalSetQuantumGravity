```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
def a(t,s):
    return np.sqrt(((t**2)/4) + (s**2))
def b(t,s):
    return ((t**2)-(s**2))/(2*t)
def delt(t,s):
    return(((t**2)+(s**2))/(2*t))*((t/2)-b(t,s))
def Ar(t,s):
    if s <= t/2:
        return 2*(b(t,s)-a(t,s))*s - 2*(s**2)*np.log(((b(t,s)/s) + np.sqrt(((b(t,s)/s)**2)+1))/((a(t,s)/s) + np.sqrt(((a(t,s)/s)**2)+1)))+(2*np.sqrt(((s*a(t,s))**2)+(a(t,s)**4)))-(2*np.sqrt(((s*b(t,s))**2)+(b(t,s)**4)))+delta(t,s)
    else:
        return 0.0
    
    
    
def Arintegrand(x,t,s):

    return 2.0*(t-(2.0*np.sqrt((x**2)+(s**2))))
    
def area(t,s):
    return quad(Arintegrand,a(t,s),b(t,s),args=(t,s))[0] + delt(t,s)
    
print area(10,1), Ar(10,1)
    
    
    
```

```python
slist = np.linspace(1,20,100)
plt.figure()
results = [Ar(5,i) for i in slist]
plt.xlabel('sigma')
plt.ylabel('Area')
plt.plot(slist,results)
plt.plot(slist,slist)
plt.show()
```

```python
def number(rho):
    return rho*Ar(100,1.0/np.sqrt(rho))
results = np.linspace(1,1000,100)
plt.figure()
plt.xlabel('rho')
plt.ylabel('number of causal set elements')
plt.plot(results,[number(i) for i in results])
plt.show()
```

```python
from scipy.integrate import quad

k = 1/np.sqrt(2)



def nintegrand(x,rho):
    return (1/(k-(2*x)))*(np.exp(-3*rho*x*k) - np.exp(-rho*k*(x+k)))




def eta(rho):
    return np.abs(quad(nintegrand,0.0000000000001, k,args=(rho))[0]/(rho*area(1.0,1/np.sqrt(rho))))


#plt.figure()
#r = np.linspace(1,1000,1000)
#plt.plot(r,[quad(nintegrand,0.00000000000001,1/np.sqrt(2),args=(i))[0] for i in r])
#plt.xlabel('rho')
#plt.ylabel('<n>')
#plt.show()



plt.figure()
r = np.linspace(2.1,100,100)
plt.xlabel(r'$\rho$')
plt.ylabel(r'$\eta, <n>, \rho A$')
plt.plot(r,[eta(i) for i in r],label=r'$\eta$')
plt.plot(r,[quad(nintegrand,0.0000000000001, k,args=(i))[0] for i in r],label='<n>')
plt.plot(r,[(i*area(1.0,1/np.sqrt(i))) for i in r],label=r'$\rho A$')
plt.yscale('log')
plt.legend()
plt.show()


```

```python
#Finds sigma for which Area is 1/sqrt(rho) (what we want)

from scipy.optimize import root


#Routine to get plot against rho of optimal sigma value
rho = np.linspace(100,1000,100)
tau = 100 #fixed

diffresults = []

for i in rho:

    def diff(s):
        return area(tau,s)-(1/np.sqrt(i))
    
    diffresults.append(root(diff,0.1).x[0])
    
#sigma required for A =1/root(rho) for varying rho and fixed tau


plt.figure()
plt.plot(rho,diffresults)
plt.show()
#plt.figure()
#slist = np.linspace(0.00001,1.0,100)
#plt.plot(slist,[area(tau,i) for i in slist]-(1/np.sqrt(rho)))
#plt.show()



#sigma required for A =1/root(rho) for varying tau and fixed rho
rho = 100
tau = np.linspace(1,100,100)
diffresults=[]
for i in tau:
    def diff(s):
        return area(i,s) - (1/np.sqrt(rho))
    diffresults.append(root(diff,0.1).x[0])
plt.figure()
plt.plot(tau,diffresults)
plt.show()   


```

```python
#is there a value of sigma for which <n> ~ rhoA? (hopefully!)
#This is a test of the validity of the continuum approximation


rho = 1 #fixed
tau = 10 #fixed
k = tau/np.sqrt(2)
def nintegrand(x,rho):
    return (1/(k-(2*x)))*(np.exp(-3*rho*x*k) - np.exp(-rho*k*(x+k)))
def difference(s):
    #print (rho*area(tau,s)), quad(nintegrand,0.000000000000001,k,args=(rho))[0]
    return (rho*area(tau,s)) -quad(nintegrand,0.000000000000001,k,args=(rho))[0]

s = np.logspace(-5,0.5,1000)
plt.figure()
plt.plot(s,[difference(i) for i in s])
plt.xscale('log')
plt.show()

root(difference,0.01).x[0]
```

```python
import matplotlib

font = {'family' : 'normal',
        'weight': 'normal',
        'size'   : 22}

matplotlib.rc('font', **font)


rholist = np.linspace(1,100,100) #fixed
results = []
for i in rholist:
    tau = 10 #fixed
    k = tau/np.sqrt(2)
    def nintegrand(x,rho):
        return (1/(k-(2*x)))*(np.exp(-3*rho*x*k) - np.exp(-rho*k*(x+k)))
    def difference(s):
        #print (rho*area(tau,s)), quad(nintegrand,0.000000000000001,k,args=(rho))[0]
        return (i*area(tau,s)) -quad(nintegrand,0.000000000000001,k,args=(i))[0]

    results.append(root(difference,0.01).x[0])
    
plt.figure()
plt.plot(rholist,results,label=r'$\sigma$')
plt.plot(rholist,[1.0/(i**2) for i in rholist],label=r'$\frac{1}{\sqrt{\rho}}$')
#plt.xscale('log')
#plt.yscale('log')
plt.legend()
plt.xlabel(r'$\rho$')
plt.ylabel(r'$\sigma$')
plt.show()

plt.figure()
plt.ylabel(r'$\eta$')
plt.xlabel(r'$\rho$')
plt.plot(rholist,[np.abs(results[i]-(1/(rholist[i]**2))) for i in range(len(rholist))])
#plt.xscale('log')
#plt.yscale('log')
plt.show()
```
